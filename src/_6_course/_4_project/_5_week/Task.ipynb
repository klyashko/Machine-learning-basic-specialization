```python
import requests
import bs4
```

```python
request = requests.get('https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff')
soup = bs4.BeautifulSoup(request.text, 'lxml')
```

```python
request.text
```

```python
for t in soup.find_all('span', attrs={'class': 'toctext'}):
    print(t.text)
```

```python
request = requests.get('https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms')
soup = bs4.BeautifulSoup(request.text, 'lxml')
rs = soup.find_all('div', attrs={'class':'mw-category-group'})
```

```python
for tag in rs:
    print(tag.find('h3').text)
    print()
    for x in tag.find_all('a'):
        if 'title' in x.attrs:
            print(x.attrs['title'])
            
    print('--------------------------------------------------------------------------')
    print()
```
