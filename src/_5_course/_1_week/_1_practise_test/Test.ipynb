```python
import pandas as pd
import numpy as np

import statsmodels.api as sm

%pylab inline
```

```python
data = pd.read_csv('monthly-milk-production.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
data.head()
```

```python
data.plot()
```

```python
_, p, _, _, _, _ = sm.tsa.stattools.adfuller(data.milk)
```

```python
print(round(p, 2))
```

```python
average_per_day = data.milk / data.index.days_in_month
print(round(sum(average_per_day), 2))
```

```python
data['per_day'] = average_per_day
data.plot(y='per_day')
```
