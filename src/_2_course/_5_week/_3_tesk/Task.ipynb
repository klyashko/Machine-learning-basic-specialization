```python
import sys
import numpy as np
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score

sys.path.append('.\\..\\..\\..\\..')

from src.utils.Utils import save_answer
```

```python
data = load_digits()

to_test = int(len(data.target) * 0.25)
X_train = data.data[:-to_test]
X_test = data.data[-to_test:]
y_train = data.target[:-to_test]
y_test = data.target[-to_test:]

print(len(y_train))
print(len(y_test))
```

```python
def distance(X1, X2):
    return np.sum((X1 - X2) ** 2)

predicts = []

for y_idx, y in enumerate(X_test):
    temp = []
    for x_idx, x in enumerate(X_train):
        dist = distance(x, y)
        temp.append((dist, y_train[x_idx]))
        
        if x_idx % 100 == 0:
            print('X_test itter ', y_idx, ' X_train itter ', x_idx)
    
    temp.sort()
    predicts.append(temp[0][1])
    
ans1 = 1 - accuracy_score(y_test, predicts)

save_answer(1, ans1)
        
```
